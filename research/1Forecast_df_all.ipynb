#Найдем прогнозные значения на интервале с 2023-11-11 по 2024-01-09

На этом интервале реальных данных уже нет

```python
# Преобразуем столбец 'date' в формат datetime (если он еще не в этом формате)
#df_real2['date'] = pd.to_datetime(df_real2['date'])
df_lstm['date'] = pd.to_datetime(df_lstm['date'])
df_prophet2['date'] = pd.to_datetime(df_prophet2['date'])
df_fourier2['date'] = pd.to_datetime(df_fourier2['date'])
df_mean['date'] = pd.to_datetime(df_mean['date'])

# Зададим столбец 'date' в качестве индекса
#df_real2.set_index('date', inplace=True)
df_lstm.set_index('date', inplace=True)
df_prophet2.set_index('date', inplace=True)
df_fourier2.set_index('date', inplace=True)
df_mean.set_index('date', inplace=True)

# Выделим нужный диапазон дат
start_date = '2023-11-11'
end_date = '2024-01-09'

#df_real2 = df_real2.loc[start_date:end_date]
df_lstm = df_lstm.loc[start_date:end_date]
df_prophet2 = df_prophet2.loc[start_date:end_date]
df_fourier2 = df_fourier2.loc[start_date:end_date]
df_mean = df_mean.loc[start_date:end_date]

```

```python
# Возможно, вам будет нужно сбросить индекс, если вы собираетесь использовать 'date' в дальнейшем как столбец:
#df_real2.reset_index(inplace=True)
df_lstm.reset_index(inplace=True)
df_prophet2.reset_index(inplace=True)
df_fourier2.reset_index(inplace=True)
df_mean.reset_index(inplace=True)
```

```python
# Устанавливаем 'date' в качестве индекса для каждого датафрейма
#df_real2.set_index('date', inplace=True)
df_lstm.set_index('date', inplace=True)
df_prophet2.set_index('date', inplace=True)
df_fourier2.set_index('date', inplace=True)
df_mean.set_index('date', inplace=True)

```

```python
# Объединяем все датафреймы
df_all = pd.concat([df_lstm, df_prophet2, df_fourier2, df_mean], axis=1, join='inner')

# Восстанавливаем индекс
df_all.reset_index(inplace=True)

# Присваиваем имена столбцам датафрейма
df_all.columns = ['date',  'y_lstm', 'y_prophet2', 'y_fourier2', 'y_mean']
```
